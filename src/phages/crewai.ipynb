```python
import os
from crewai import Agent, Task, Crew, Process


# You can choose to use a local model through Ollama for example.
#
# from langchain.llms import Ollama
# ollama_llm = Ollama(model="openhermes")

# Install duckduckgo-search for this example:
# !pip install -U duckduckgo-search

from langchain.tools import DuckDuckGoSearchRun
search_tool = DuckDuckGoSearchRun()

# Define your agents with roles and goals
researcher = Agent(
  role='Senior Research Analyst',
  goal='Uncover cutting-edge developments in AI and data science',
  backstory="""You work at a leading tech think tank.
  Your expertise lies in identifying emerging trends.
  You have a knack for dissecting complex data and presenting
  actionable insights.""",
  verbose=True,
  allow_delegation=False,
  tools=[search_tool]
  # You can pass an optional llm attribute specifying what mode you wanna use.
  # It can be a local model through Ollama / LM Studio or a remote
  # model like OpenAI, Mistral, Antrophic of others (https://python.langchain.com/docs/integrations/llms/)
  #
  # Examples:
  # llm=ollama_llm # was defined above in the file
  # llm=ChatOpenAI(model_name="gpt-3.5", temperature=0.7)
)
writer = Agent(
  role='Tech Content Strategist',
  goal='Craft compelling content on tech advancements',
  backstory="""You are a renowned Content Strategist, known for
  your insightful and engaging articles.
  You transform complex concepts into compelling narratives.""",
  verbose=True,
  allow_delegation=True,
  # (optional) llm=ollama_llm
)

# Create tasks for your agents
task1 = Task(
  description="""Conduct a comprehensive analysis of the latest advancements in AI in 2024.
  Identify key trends, breakthrough technologies, and potential industry impacts.
  Your final answer MUST be a full analysis report""",
  agent=researcher
)

task2 = Task(
  description="""Using the insights provided, develop an engaging blog
  post that highlights the most significant AI advancements.
  Your post should be informative yet accessible, catering to a tech-savvy audience.
  Make it sound cool, avoid complex words so it doesn't sound like AI.
  Your final answer MUST be the full blog post of at least 4 paragraphs.""",
  agent=writer
)

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[researcher, writer],
  tasks=[task1, task2],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)

# Get your crew to work!
result = crew.kickoff()

print("######################")
print(result)
```

```python
import os
from crewai import Agent, Task, Crew, Process


from langchain.tools import DuckDuckGoSearchRun, PubmedQueryRun, ArxivQueryRun, HumanInputRun
from langchain_experimental.tools import PythonAstREPLTool
from langchain.chat_models import ChatOpenAI

search_tool = DuckDuckGoSearchRun()
pubmed_tool = PubmedQueryRun()
arxiv_tool = ArxivQueryRun()
python_tool = PythonAstREPLTool()
human_tool = HumanInputRun()

microbiologist = Agent(
  role='Microbiologist',
  goal='Summarize scientific research on phages with the help of your team.',
  backstory="""You are a renowned microbiologist whose mission is to develop the phage directory, 
  a place where doctors can find summaries of the research on each phage that has been studied to fight antibiotic resistant bacteria in human.""",
  verbose=True,
  allow_delegation=True,
  tools=[pubmed_tool, search_tool, human_tool],
  llm=ChatOpenAI(model_name="gpt-4", temperature=0.7)
)

ai_researcher = Agent(
  role='AI Researcher',
  goal='Keep up to date with SOTA methods to summarize scientific papers using LLMs in 2023 and 2024.',
  backstory="""You are a renowned AI Researcher and worked with paperQA https://github.com/whitead/paper-qa to develop Wikicrow https://github.com/Future-House/WikiCrow/blob/main/README.md.""",
  verbose=True,
  allow_delegation=True,
  tools=[arxiv_tool, search_tool]
  # (optional) llm=ollama_llm
)

# Define your agents with roles and goals
senior_ai_engineer = Agent(
  role='Senior AI Engineer',
  goal='Design complex RAG Systems',
  backstory="""You are a top AI engineer of Llamaindex, a friend of Jerry Liu.
  Your expertise lies in setting up the right structure for your data to retrieve it with near perfect recall and precision.
  You have a knack for designing retrievers and query engines using llama_index library https://github.com/run-llama/llama_index.""",
  verbose=True,
  allow_delegation=False,
  tools=[search_tool, human_tool]
  # You can pass an optional llm attribute specifying what mode you wanna use.
  # It can be a local model through Ollama / LM Studio or a remote
  # model like OpenAI, Mistral, Antrophic of others (https://python.langchain.com/docs/integrations/llms/)
  #
  # Examples:
  # llm=ollama_llm # was defined above in the file
  # llm=ChatOpenAI(model_name="gpt-3.5", temperature=0.7)
)

ai_engineer = Agent(
  role='AI Engineer',
  goal='Code in python the design that your boss the senior_ai_engineer comes up with',
  backstory="""You are a wiz at developing RAG solution using llama_index library https://github.com/run-llama/llama_index.""",
  verbose=True,
  allow_delegation=False,
  tools=[search_tool, python_tool, human_tool]
  # You can pass an optional llm attribute specifying what mode you wanna use.
  # It can be a local model through Ollama / LM Studio or a remote
  # model like OpenAI, Mistral, Antrophic of others (https://python.langchain.com/docs/integrations/llms/)
  #
  # Examples:
  # llm=ollama_llm # was defined above in the file
  # llm=ChatOpenAI(model_name="gpt-3.5", temperature=0.7)
)



# Create tasks for your agents
task1 = Task(
  description="""Identify 3 well known bacteriophages used to kill antibiotic bacteria resistant strain. 
  Propose a document template for summarizing knowledge on each of them.
  Include a description of each section, and why this is important. 
  Finally create prompts for GPT-4 to create each of the identified sections assuming it has access to the scientific papers: you need to provide a prompt for each section
  Provide references of any scientific papers or websites used for your research.
  
  YOU MUST RETURN:
  1. THE LIST OF BACTERIOPHAGES
  2. THE SUMMARY SECTIONS DESCRIPTION
  3. THE SUMMARY PROMPTS FOR GPT-4
  
  SEEK APPROVAL FOR HUMAN BEFORE SENDING THE INFORMATION TO YOUR COLLEAGUE THE SENIOR_AI_ENGINERR.""",
  agent=microbiologist
)


task3 = Task(
  description="""Research SOTA abstractive summarization methods for multiple scientific papers on a topic: in this case the summary template provided by your colleague the microbiologist.
   Design a rag structure using llama_index library (type of index, additional metadata, retriever type) that would support the summarizing of a collection of 
  scientific papers on a phage. Don't create a basic RAG that will provide mediocre result, instead aim for great precision and recall in the retrieval. Provide references.
    YOU MUST RETURN:
    1. INDEX STRUCTURE DESIGN READY FOR AN ENGINEER TO CODE

    SEEK APPROVAL FOR HUMAN BEFORE SENDING THE INFORMATION TO YOUR COLLEAGUE THE AI_ENGINERR.
       """,
  agent=senior_ai_engineer
)

task4 = Task(
  description="""Write the code using llama_index and any other required library that allows to summarize a collection of scientific papers on a phage into the required template.""",
  agent=ai_engineer
)

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[microbiologist, senior_ai_engineer, ai_engineer],
  tasks=[task1, task3, task4],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)

# Get your crew to work!
result = crew.kickoff()

print("######################")
print(result)
```

```python
import os
from crewai import Agent, Task, Crew, Process


from langchain.tools import DuckDuckGoSearchRun, PubmedQueryRun, ArxivQueryRun, HumanInputRun
from langchain_experimental.tools import PythonAstREPLTool
from langchain.chat_models import ChatOpenAI

search_tool = DuckDuckGoSearchRun()
pubmed_tool = PubmedQueryRun()
arxiv_tool = ArxivQueryRun()
python_tool = PythonAstREPLTool()
human_tool = HumanInputRun()

microbiologist = Agent(
  role='Microbiologist',
  goal='Review scientific papers on bacteriophages killing antibiotic resistant bacteria in human.',
  backstory="""You are a renowned microbiologist whose mission is to develop the phage directory, 
  a place where doctors can find summaries of the research on each phage that has been studied to fight antibiotic resistant bacteria in human.""",
  verbose=True,
  allow_delegation=True,
  tools=[arxiv_tool, search_tool, pubmed_tool],
  llm=ChatOpenAI(model_name="gpt-4", temperature=0.7)
)

microbiologist2 = Agent(
  role='Microbiologist',
  goal='Review scientific papers on bacteriophages killing antibiotic resistant bacteria in human.',
  backstory="""You are a renowned microbiologist whose mission is to develop the phage directory, 
  a place where doctors can find summaries of the research on each phage that has been studied to fight antibiotic resistant bacteria in human.""",
  verbose=True,
  allow_delegation=True,
  tools=[arxiv_tool, search_tool, pubmed_tool],
  llm=ChatOpenAI(model_name="gpt-4", temperature=0.7)
)

microbiologist_manager = Agent(
  role='Microbiologist Manager',
  goal='Summarize and publish information returned by his team, microbiologist1 and microbiologist2.',
  backstory="""You are a renowned microbiologist whose mission is to develop the phage directory, 
  a place where doctors can find summaries of the research on each phage that has been studied to fight antibiotic resistant bacteria in human.""",
  verbose=True,
  allow_delegation=True,
  tools=[arxiv_tool, search_tool, pubmed_tool],
  llm=ChatOpenAI(model_name="gpt-4", temperature=0.7)
)


bacteriophage_name = "Phi X 174 Phage"

introduction_task_tpl = """Write a comprehensive introduction for the bacteriophage {bacteriophage_name}. 
  Include details on its discovery, historical background, and significance in the realm of antibiotic-resistant bacterial treatments. 
  Cite relevant academic journals and scientific studies to support the information.
  """

introduction_task = introduction_task_tpl.format(bacteriophage_name=bacteriophage_name)

# Create tasks for your agents
task1 = Task(
  description=introduction_task,
  agent=microbiologist
)

task2 = Task(
  description=introduction_task,
  agent=microbiologist2
)

task3 = Task(
  description=introduction_task,
  agent=microbiologist2
)

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[microbiologist],
  tasks=[task1],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)

# Get your crew to work!
result = crew.kickoff()

print("######################")
print(result)
```
